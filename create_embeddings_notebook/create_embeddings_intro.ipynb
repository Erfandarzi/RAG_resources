## Pull up text by vector similarity

```python
# Save in a vector database -- install relevant library
!pip install langchain-chroma
!pip install faiss-cpu
```

```python
# Store vectors in an index

import faiss
import numpy as np

# Assuming embeddings is a list of lists
embedding_dim = len(embeddings[0])
index = faiss.IndexFlatL2(embedding_dim)  # Use L2 distance

# Convert embeddings to a numpy array
embedding_array = np.array(embeddings).astype('float32')

# Add embeddings to the index
index.add(embedding_array)
```

```python
# Embed the query
query = "red dogs"
query_embedding = embeddings_model.embed_query(query)

# Convert the query embedding to a numpy array
query_embedding_array = np.array([query_embedding]).astype('float32')

# Search the index for the most similar documents
k = 1  # Number of nearest neighbors to retrieve
distances, indices = index.search(query_embedding_array, k)

# Retrieve the most similar documents
similar_docs = [documents[i] for i in indices[0]]
```

```python
response = similar_docs[0]
print(response)
```

## Write response based on similar text

```python
from langchain.chains import LLMChain
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
```

```python
# Step 2: Query the index
query = "I'm looking for a small red dog"
query_embedding = embeddings_model.embed_query(query)
query_embedding_array = np.array([query_embedding]).astype('float32')

k = 2  # Number of nearest neighbors to retrieve
distances, indices = index.search(query_embedding_array, k)
similar_docs = [documents[i] for i in indices[0]]

# Step 3: Combine the retrieved documents
context = " ".join(similar_docs)

# Step 4: Generate a reply using LangChain
# Initialize the language model (you might need to replace with your specific model and API key)
llm = OpenAI(temperature=0.7, openai_api_key=OPENAI_API_KEY)

# Create a prompt template
prompt_template = PromptTemplate(
    template="Given the following information, provide a one-sentence dog recommendation.\n\nContext: {context}\n\nQuery: {query}\n\nResponse:",
    input_variables=["context", "query"]
)

# Create a chain
chain = LLMChain(llm=llm, prompt=prompt_template)

# Generate the response
response = chain.run({
    "context": context,
    "query": query
})

print(response)
```
